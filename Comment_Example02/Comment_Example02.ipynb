```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("wholesale_customers_data.csv")
```

```python
data.isnull().sum()
```

```python
#Before
outliers = {}
#The following code snippet allows you to look for outliers
for i in range(data.shape[1]):
    min_t = data[data.columns[i]].mean() - (3 * data[data.columns[i]].std())
    max_t = data[data.columns[i]].mean() + (3 * data[data.columns[i]].std())
    count = 0
# The following code snippet allows you to look for outliers in the entire set of features at once
    for j in data[data.columns[i]]:
        if j < min_t or j > max_t:
            count += 1
    outliers[data.columns[i]] = [count,data.shape[0]-count]

print(outliers)
```

```python
#After
outliers = {}
#The following code snippet allows you to look for outliers
for i in range(data.shape[1]):
    min_t = data[data.columns[i]].mean() - (3 * data[data.columns[i]].std())
    max_t = data[data.columns[i]].mean() + (3 * data[data.columns[i]].std())
    count = 0
    """
    The following code snippet allows you to look for outliers in \
    the entire set of features at once
    """
    for j in data[data.columns[i]]:
        if j < min_t or j > max_t:
            count += 1
    outliers[data.columns[i]] = [count,data.shape[0]-count]
print(outliers)
```
